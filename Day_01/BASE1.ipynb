```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('f1.png')
if image is None:
    print('Image not found')
else:
    print('Image found')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
image_bgr = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_bgr, cmap='gray')
plt.title('BGR Image')
plt.axis('off')
plt.show()
cv2.imwrite('Gray.png', image_bgr)
```

```python
blur = cv2.GaussianBlur(image_bgr, (11,11), 0)

plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
plt.show()

# cv2.imwrite('blurred_face.png', blur)
# print("✅ Saved blurred image as blurred_face.png")
```

```python
blur = cv2.GaussianBlur(image_bgr, (7,7), 0)
plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
plt.show()

blur = cv2.GaussianBlur(image_bgr, (11,11), 0)
plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
plt.show()
```
